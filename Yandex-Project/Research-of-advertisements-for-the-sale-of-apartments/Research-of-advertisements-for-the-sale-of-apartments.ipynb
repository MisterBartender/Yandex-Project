**Этаж**

Число таких пропусков достаточно не большое, но данные необходимы в анализе. Можно замениить на медиану высоты в населённом пункте

```python
# заменяем пропуски на 0, т.к. такого этажа нет в данных, и будет удобно работать с целым числом
df['floors_total'] = df['floors_total'].fillna(0)
# меняем формать ячейки на целочисленный
df['floors_total'] = df['floors_total'].astype('int')
# переменая для медианого значения по локации объекта
floors_total_median = dict(df.pivot_table(index = 'locality_name', values = 'floors_total', aggfunc = 'median')['floors_total'])
type(floors_total_median)
# функция для измениения значения 0 на этажность по населенному пункту
def floors_total_fun(row):
    loc_name = row['locality_name']
    fl_total = row['floors_total']
    if fl_total == 0:
        for key, value in floors_total_median.items():
            if key in loc_name:
                return value
    
    return fl_total
# внесение изменения в таблицу
df['floors_total'] = df.apply(floors_total_fun, axis=1)
```

Посчитаем цену квадратного метра относительно общей площади, добамим соответствующие значения в таблицу и приведем к целочисленному значению, для удобства просмотра

```python
# день недели публикации
df['weekday_exposition'] = df['first_day_exposition'].dt.day_name()
```

```python
#месяц публикации
df['month_exposition'] = df['first_day_exposition'].dt.month_name()
```

```python
#год публикации
df['year_exposition'] = df['first_day_exposition'].dt.year
```

### Создади категории этажей (первый, последний, другой)

```python
# функция для присваивания категории этажа
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
```

```python
# присваивание категории этажа
df['floor_category'] = df.apply(floor_category, axis = 1)
```
